```python
import pandas as pd
```

```python
last_season = pd.read_csv('2022-23.csv')
season_prior = pd.read_csv('2021-22.csv')
two_years_ago = pd.read_csv('2020-21.csv')
```

```python
last_three = pd.concat([last_season, season_prior, two_years_ago], axis=0)
```

```python
last_three = last_three.loc[last_three['position'] != "G"]
```

```python
players_list = last_three['name'].unique()
```

```python
players_list
```

```python
noPoints_df = pd.DataFrame(columns=['Name', 'Position', '2021GP', '2021FP', '2021NP', '2021TOI', '2122GP', '2122FP', '2122NP', '2122TOI', '2223GP', '2223FP', '2223NP', '2223TOI'])

for player in players_list:
    current = last_three.loc[last_three['name'] == player]
    position = current['position'].iloc[0]
    s2021 = current.loc[(current['date'] >= '2021-01-01') & (current['date'] <= '2021-08-01')]
    s2122 = current.loc[(current['date'] >= '2021-10-01') & (current['date'] <= '2022-08-01')]
    s2223 = current.loc[(current['date'] >= '2022-10-01') & (current['date'] <= '2023-08-01')]

    gp2021 = len(s2021)
    gp2122 = len(s2122)
    gp2223 = len(s2223)
    
    fp2021 = s2021['fantasyPoints'].sum()
    fp2122 = s2122['fantasyPoints'].sum()
    fp2223 = s2223['fantasyPoints'].sum()
    
    bs2021 = s2021['blocked'].sum()
    bs2122 = s2122['blocked'].sum()
    bs2223 = s2223['blocked'].sum()
    
    h2021 = s2021['hits'].sum()
    h2122 = s2122['hits'].sum()
    h2223 = s2223['hits'].sum()
    
    sg2021 = s2021['shots'].sum()
    sg2122 = s2122['shots'].sum()
    sg2223 = s2223['shots'].sum()
    
    np2021 = (bs2021*.5) + ((h2021 + sg2021)*.1)
    np2122 = (bs2122*.5) + ((h2122 + sg2122)*.1)
    np2223 = (bs2223*.5) + ((h2223 + sg2223)*.1)
    
    toi2021 = s2021['timeOnIce'].sum()
    toi2122 = s2122['timeOnIce'].sum()
    toi2223 = s2223['timeOnIce'].sum()
    
    current_df = {'Name': player, 'Position': position, '2021GP': gp2021, '2021FP': fp2021, '2021NP': np2021, '2021TOI': toi2021, '2122GP': gp2122, '2122FP': fp2122, '2122NP': np2122, '2122TOI': toi2122, '2223GP': gp2223, '2223FP': fp2223, '2223NP': np2223, '2223TOI': toi2223}
    
    current_df['2021BS'] = bs2021
    current_df['2122BS'] = bs2122
    current_df['2223BS'] = bs2223
    
    current_df['2021H'] = h2021
    current_df['2122H'] = h2122
    current_df['2223H'] = h2223
    
    current_df['2021S'] = sg2021
    current_df['2122S'] = sg2122
    current_df['2223S'] = sg2223
    
    current_df = pd.DataFrame([current_df])
    
    noPoints_df = pd.concat([noPoints_df, current_df], ignore_index=True)
```

```python
# noPoints_df = noPoints_df.loc[noPoints_df['2223GP'].astype(int) >= 10]
# noPoints_df = noPoints_df.loc[noPoints_df['2021GP'].astype(int) >= 10]
# noPoints_df = noPoints_df.loc[noPoints_df['2122GP'].astype(int) >= 10]


noPoints_df['2021GP'] = noPoints_df['2021GP'].astype(int)
noPoints_df['2122GP'] = noPoints_df['2122GP'].astype(int)
noPoints_df['2223GP'] = noPoints_df['2223GP'].astype(int)

noPoints_df['2021TOI'] = round(noPoints_df['2021TOI'].astype(int)/60, 2)
noPoints_df['2122TOI'] = round(noPoints_df['2122TOI'].astype(int)/60, 2)
noPoints_df['2223TOI'] = round(noPoints_df['2223TOI'].astype(int)/60, 2)




noPoints_df['2223NP_60'] = round(noPoints_df['2223NP']/noPoints_df['2223TOI'] * 60, 2)
noPoints_df['2122NP_60'] = round(noPoints_df['2122NP']/noPoints_df['2122TOI'] * 60, 2)
noPoints_df['2021NP_60'] = round(noPoints_df['2021NP']/noPoints_df['2021TOI'] * 60, 2)

noPoints_df['2223NP_GP'] = round(noPoints_df['2223NP']/noPoints_df['2223GP'], 2)
noPoints_df['2122NP_GP'] = round(noPoints_df['2122NP']/noPoints_df['2122GP'], 2)
noPoints_df['2021NP_GP'] = round(noPoints_df['2021NP']/noPoints_df['2021GP'], 2)

noPoints_df['2223FPP60'] = round(noPoints_df['2223FP']/noPoints_df['2223TOI'] * 60, 2)
noPoints_df['2122FPP60'] = round(noPoints_df['2122FP']/noPoints_df['2122TOI'] * 60, 2)
noPoints_df['2021FPP60'] = round(noPoints_df['2021FP']/noPoints_df['2021TOI'] * 60, 2)

noPoints_df['2223FPPG'] = round(noPoints_df['2223FP']/noPoints_df['2223GP'], 2)
noPoints_df['2122FPPG'] = round(noPoints_df['2122FP']/noPoints_df['2122GP'], 2)
noPoints_df['2021FPPG'] = round(noPoints_df['2021FP']/noPoints_df['2021GP'], 2)

noPoints_df['2223NP%'] = round(noPoints_df['2223NP']/noPoints_df['2223FP']*100, 2)
noPoints_df['2122NP%'] = round(noPoints_df['2122NP']/noPoints_df['2122FP']*100, 2)
noPoints_df['2021NP%'] = round(noPoints_df['2021NP']/noPoints_df['2021FP']*100, 2)
```

```python
noPoints_df['TOT_GAMES'] = noPoints_df['2223GP'] + noPoints_df['2122GP'] + noPoints_df['2021GP']

noPoints_df['TOT_NP_60'] = round((noPoints_df['2223NP'] + noPoints_df['2122NP'] + noPoints_df['2021NP']) / (noPoints_df['2223TOI'] + noPoints_df['2122TOI'] + noPoints_df['2021TOI']) * 60, 2)
noPoints_df['TOT_NP_GP'] = round((noPoints_df['2223NP'] + noPoints_df['2122NP'] + noPoints_df['2021NP']) / (noPoints_df['2223GP'] + noPoints_df['2122GP'] + noPoints_df['2021GP']), 2)
noPoints_df['TOT_FPP60'] = round((noPoints_df['2223FP'] + noPoints_df['2122FP'] + noPoints_df['2021FP']) / (noPoints_df['2223TOI'] + noPoints_df['2122TOI'] + noPoints_df['2021TOI']) * 60, 2)
noPoints_df['TOT_FPPG'] = round((noPoints_df['2223FP'] + noPoints_df['2122FP'] + noPoints_df['2021FP']) / (noPoints_df['2223GP'] + noPoints_df['2122GP'] + noPoints_df['2021GP']), 2)
noPoints_df['TOT_NP%'] = round((noPoints_df['2223NP'] + noPoints_df['2122NP'] + noPoints_df['2021NP']) / (noPoints_df['2223FP'] + noPoints_df['2122FP'] + noPoints_df['2021FP'])*100, 2)

noPoints_df = noPoints_df.loc[noPoints_df['TOT_GAMES'] >= 20]
noPoints_df
```

```python
noPoints_df['LAST2_GAMES'] = noPoints_df['2223GP'] + noPoints_df['2122GP']

noPoints_df['LAST2_NP_60'] = round((noPoints_df['2223NP'] + noPoints_df['2122NP']) / (noPoints_df['2223TOI'] + noPoints_df['2122TOI']) * 60, 2)
noPoints_df['LAST2_NP_GP'] = round((noPoints_df['2223NP'] + noPoints_df['2122NP']) / (noPoints_df['2223GP'] + noPoints_df['2122GP']), 2)
noPoints_df['LAST2_FPP60'] = round((noPoints_df['2223FP'] + noPoints_df['2122FP']) / (noPoints_df['2223TOI'] + noPoints_df['2122TOI']) * 60, 2)
noPoints_df['LAST2_FPPG'] = round((noPoints_df['2223FP'] + noPoints_df['2122FP']) / (noPoints_df['2223GP'] + noPoints_df['2122GP']), 2)
noPoints_df['LAST2_NP%'] = round((noPoints_df['2223NP'] + noPoints_df['2122NP']) / (noPoints_df['2223FP'] + noPoints_df['2122FP'])*100, 2)


noPoints_df
```
